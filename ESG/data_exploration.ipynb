```python
import pandas as pd
import os

# Construct the file path using os.path.join
file_path = os.path.join("EUESGMANUFACTURER.csv")
try:
    # Load the CSV file
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print(f"The file {file_path} was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# Specify the column name you're interested in
column_name = ["companyDomicileISO", "companyLongName", "ESGActivity", "ESGClassification","ESGFactor"]

for i in column_name:
        # Extract unique entries for the specified column
    unique_entries = df[i].unique()
    
    # Convert the unique entries to a list (optional, as unique() already returns an array)
    unique_entries_list = list(unique_entries)
    
    print(f"Unique entries in '{i}':", unique_entries_list)
    print(len(unique_entries_list))
```
